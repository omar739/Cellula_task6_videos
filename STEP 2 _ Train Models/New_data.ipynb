```python
import os 
import numpy as np
import cv2
import time
```

## **Convert Shop_videos and save them to new dir**

```python
path_shop = "D:/DEEP Leaning dataset/shop videos/Shop DataSet/shop lifters"
path_shop_new = "D:/DEEP Leaning dataset/shop videos/Shop DataSet/shop_edited"
videos = os.listdir(path_shop)
start = time.time()
for i in range(len(videos)) :
    vid_path = os.path.join(path_shop,videos[i])
    video  = cv2.VideoCapture(vid_path)
    #---------------------------------------------------------
    # Get videos information ( width , height , frames , FPS )
    #---------------------------------------------------------
    width  = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = video.get(cv2.CAP_PROP_FPS)
    frames = video.get(cv2.CAP_PROP_FRAME_COUNT)
    #-----------------------------------------------
    # Change ( FPS = 25 ) ( number_of_frames = 70 )
    #-----------------------------------------------
    FPS = 25
    path_of_save = os.path.join(path_shop_new,videos[i])
    out_video = cv2.VideoWriter(path_of_save, cv2.VideoWriter_fourcc(*'mp4v'), 
                                FPS , isColor=True, frameSize=(width,height))
    step = int(frames /70)
    frame_index = 0
    frame_Written = 0
    while video.isOpened() and frame_Written < 70 :
        video.set(cv2.CAP_PROP_POS_FRAMES,frame_index)
        ret , frame = video.read()
        if not ret :
            break
        out_video.write(frame)
        frame_index += step
        frame_Written +=1
    out_video.release()
    video.release()
end = time.time()
print(f"Done in {end-start} sec")
```

```python
#Chech now if it changed or not
vid_edited = os.listdir(path_shop_new)
for i in vid_edited :
    pat = os.path.join(path_shop_new,i)
    vid = cv2.VideoCapture(pat)
    #-----------------------------------------------
    width  = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = vid.get(cv2.CAP_PROP_FPS)
    frames = vid.get(cv2.CAP_PROP_FRAME_COUNT)
    #----------------------------------------------
    print(f" for {i} after edit , width = {width} , height = { height} , FPS = {fps} , frames = {frames}")
```
